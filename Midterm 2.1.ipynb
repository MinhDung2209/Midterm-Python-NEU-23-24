```python
# Import các thư viện cần thiết
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
import numpy as np

# Đọc dữ liệu
file_path = r"C:\Users\Admin\OneDrive - National Economics University\Documents\Học liệu\Junior - Third year\Semester 1\Python Programing\winequality-red.csv"
wine_data = pd.read_csv(file_path)

# Xem qua dữ liệu
wine_data.head(), wine_data.columns
```

```python
# Tạo biểu đồ scatter plot với fixed acidity và quality
plt.figure(figsize=(10, 6))
sns.scatterplot(x=wine_data['fixed acidity'], y=wine_data['quality'], color="blue", label="Data points")

# Thêm đường hồi quy tuyến tính
X = wine_data['fixed acidity'].values.reshape(-1, 1)
y = wine_data['quality'].values
reg = LinearRegression().fit(X, y)
plt.plot(X, reg.predict(X), color="red", label=f"Regression Line: y = {reg.coef_[0]:.2f}x + {reg.intercept_:.2f}")

plt.xlabel("Fixed Acidity")
plt.ylabel("Quality")
plt.title("Linear Regression: Fixed Acidity vs Quality")
plt.legend()
plt.show()
```

```python
# Tạo biểu đồ scatter plot với pH và quality
plt.figure(figsize=(10, 6))
sns.scatterplot(x=wine_data['pH'], y=wine_data['quality'], color="blue", label="Data points")

# Thêm đường hồi quy tuyến tính
X = wine_data['pH'].values.reshape(-1, 1)
y = wine_data['quality'].values
reg = LinearRegression().fit(X, y)
plt.plot(X, reg.predict(X), color="red", label=f"Regression Line: y = {reg.coef_[0]:.2f}x + {reg.intercept_:.2f}")

plt.xlabel("pH")
plt.ylabel("Quality")
plt.title("Linear Regression: pH vs Quality")
plt.legend()
plt.show()

```

```python
# Biểu đồ 3D để hiển thị Fixed Acidity, pH và Quality
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(wine_data['fixed acidity'], wine_data['pH'], wine_data['quality'], color='purple', marker='o')
ax.set_xlabel('Fixed Acidity')
ax.set_ylabel('pH')
ax.set_zlabel('Quality')
ax.set_title('3D Relationship: Fixed Acidity, pH, and Quality')
plt.show()
```

```python
# Bản đồ nhiệt để xem mối tương quan giữa các yếu tố
plt.figure(figsize=(8, 6))
correlation_matrix = wine_data[['fixed acidity', 'pH', 'quality']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", square=True)
plt.title("Correlation Heatmap")
plt.show()
```
